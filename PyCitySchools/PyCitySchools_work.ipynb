```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# count the distinct number of schools in the district
total_schools = school_data_complete['school_name'].nunique()
total_schools
```

```python
# count the total number of students in the district
total_students = school_data_complete['student_name'].count()
total_students
```

```python
# group by school to determine the total budget
school_group = school_data_complete.groupby(['school_name'])
# take the average budget by school to account for duplicate values
school_mean = school_group.mean()
# take the sum of the newly calculated average budget of each school to determine the total budget
total_budget = school_mean['budget'].sum()
total_budget
```

```python
# determine the average math score for the district
avg_math = school_data_complete['math_score'].mean()
avg_math
```

```python
# determine the average reading score for the district
avg_reading = school_data_complete['reading_score'].mean()
avg_reading
```

```python
# Create bins/labels to determine % of passing math, reading, and overall passing scores
bins = [0,69.9,100]
grade = ['Failing', 'Passing']
```

```python
# % of passing math scores
school_data_complete['passing_math'] = pd.cut(school_data_complete['math_score'], bins, labels=grade)
passing_math = school_data_complete['passing_math'].value_counts()
percent_passing_math = passing_math['Passing'] / total_students * 100
percent_passing_math
```

```python
# % of passing reading scores
school_data_complete['passing_reading'] = pd.cut(school_data_complete['reading_score'], bins, labels=grade)
passing_reading = school_data_complete['passing_reading'].value_counts()
percent_passing_reading = passing_reading['Passing'] / total_students * 100
percent_passing_reading
```

```python
# % overall passing
passing_overall= school_data_complete['passing_overall'] = school_data_complete.loc[((school_data_complete['passing_math'] == 'Passing') & (school_data_complete['passing_reading'] == 'Passing')), 'student_name'].count()
passing_overall
percent_overall = passing_overall / total_students * 100
percent_overall
```

```python
school_data_complete
```

```python
# Create a dataframe to hold results:

district_summary_df = pd.DataFrame( 
    { 'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [avg_math],
    'Average Reading Score': [avg_reading],
    '% Passing Math': [percent_passing_math],
    '% Passing Reading': [percent_passing_reading],
    '% Passing Overall': [percent_overall]}
                                  )
district_summary_df
```

```python
#create School Summary fields
school_name = school_data.loc[:, 'school_name']
school_name
```

```python
school_type = school_data.loc[:, 'type']
school_type
```

```python
# calculate total students per school
total_students_by_school = school_data['size']
total_students_by_school
```

```python
#calculate total school budget
school_budget = school_data['budget']
school_budget
```

```python
#calculate total school budget per student

per_student_budget = school_budget / total_students_by_school
per_student_budget
```

```python
# calculate average scores for reading and math

student_group = pd.DataFrame(student_data.groupby('school_name').mean())
avg_math_school = student_group.loc[:, 'math_score'].count()
avg_reading_school = student_group.loc[:, 'reading_score'].count()
```

```python
# calculate the amount of students passing math, reading, and passing overall by school
school_data_complete['math_passing_school'] = school_data_complete.loc[school_data_complete['math_score'] >= 70, 'student_name']
school_data_complete['reading_passing_school'] = school_data_complete.loc[school_data_complete['reading_score'] >= 70, 'student_name']
school_data_complete['overall_passing_school'] = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70), 'student_name']
```

```python
# groupings for percentage calculations
school_group = school_data_complete.groupby('school_name').count()
students_by_school = school_group.loc[:, 'size']
group_math_pass = school_group.loc[:,'math_passing_school']
group_reading_pass = school_group.loc[:, 'reading_passing_school']
group_overall_pass = school_group.loc[:, 'overall_passing_school']
```

```python
#percent passing
percent_math_passing_school = group_math_pass / students_by_school * 100
percent_reading_passing_school = group_reading_pass / students_by_school * 100
percent_passing_overall_school = group_overall_pass / students_by_school * 100
```

```python
# School Summary Data Frame

school_summary_df = pd.DataFrame(
    {'school_name': school_name,
     'School Type': school_type,
     'Total Students': total_students_by_school,
     'Total School Budget': school_budget,
     'Per Student Budget': per_student_budget,}
)


school_summary_second_df = pd.DataFrame(
    {'% Passing Math': percent_math_passing_school,
   '% Passing Reading': percent_reading_passing_school,
    '% Passing Overall': percent_passing_overall_school,}
)
```

```python
#begin merging dataframes
df_merge =  pd.merge(school_summary_df, student_group, on='school_name')
df_merge
```

```python
school_summary_complete_df = pd.merge(df_merge, school_summary_second_df, on='school_name')
school_summary_complete_df.set_index('school_name', inplace=True)
```

```python
school_summary_complete_df.rename(columns={'reading_score': "Average Reading Score", 'math_score': 'Average Math Score'}, inplace=True)
school_summary_complete_df.drop(columns='Student ID', inplace=True)
school_summary_complete_df
```
